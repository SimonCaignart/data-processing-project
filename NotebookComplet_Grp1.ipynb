&emsp; &emsp;Notre projet est un problème d’étude connu sous le nom de problème de tournée de véhicule, aussi appelé VRP pour Vehicle Routing Problem exposé dans [[1]](https://www.lgi2a.univ-artois.fr/spip/IMG/pdf/these_zhao_xin.pdf).   

Imaginez-vous en train de commander des produits sur internet, une fois la confirmation de l’achat effectué vient la phase d’acheminement. Mais comment votre produit fait-il pour arriver à bon port et dans les temps ?    
Pleins d’éléments rentrent en jeux avant qu’il vous soit livré car le livreur doit également assurer la livraison d’autres clients. Les tournées sont donc organisées de façon à optimiser la distance totale parcourue par votre livreur, diminuer votre temps d'attente tout en prenant en compte la circulation proche de chez vous.

###### &emsp; &emsp; &emsp; **Problème d'étude connu - Problème TSP**

&emsp; &emsp;Notre projet pourrait donc se référer à un problème d’étude connu sous le nom de problème de voyageur de commerce, aussi appelé TSP pour Traveling Salesman Problem exposé dans [[2]](https://www.datavis.fr/index.php?page=salesman-problem).


Imaginez-vous en train de commander des produits sur internet, une fois la confirmation de l’achat effectué vient la phase d’acheminement. Mais comment votre produit fait-il pour arriver à bon port et dans les temps ?    
Pleins d’éléments rentrent en jeux avant qu’il vous soit livré car le livreur doit également assurer la livraison d’autres clients dans différentes villes. Les tournées sont donc organisées de façon à optimiser la distance totale parcourue par votre livreur.    
Mais ce n’est pas tout, le livreur doit terminer sa tournée dans sa ville de départ. Le livreur doit donc respecter son planning de livraisons pour pouvoir acheminer les colis des clients dans les temps et en passant une seule est unique fois dans chaque ville lors de sa tournée. 


#### &emsp; &emsp; **2.1.2 Définition mathématique du problème** <a name="2.1.2"></a>

Nous pouvons d’abord définir l’objectif de ce problème, en effet, le but de celui-ci est de trouver le trajet le plus court reliant plusieurs villes, afin de respecter au mieux l’environnement.
À partir de cet objectif, on peut définir un problème de décision, mais tout d’abord, il faut définir les données d’entrée pour ce problème de décision.

On peut imaginer en entrée un graphe complet G, comportant un ensemble de sommets V, et un ensemble d’arêtes A, chaque sommet représentant une ville et chaque arête représentant une route. On peut également ajouter un entier k représentant la distance ou le temps du trajet.
À partir de ces données et de cet objectif, on peut décrire un problème de décision :

***Existe-t-il un cycle hamiltonien dans G de longueur inférieure à k ?***

Un cycle hamiltonien est un chemin qui passe par tous les sommets d’un graphe une seule et unique fois, en revenant à son point de départ. Dans notre cas, ce cycle représente donc le trajet d’un camion qui doit passer par toutes les villes d’un sous-ensemble de villes données.

De ce fait, notre problème d’optimisation est le suivant :

***Quelle est la plus petite valeur de k pour laquelle la réponse est oui ?***

Dans une vision générale, on mesure la complexité d’un algorithme par le nombre d’opérations élémentaires. Pour commencer notre complexité du problème, nous pouvons parler de « Complexité de temps ». En effet, on doit déterminer la complexité d’un traitement en fonction de la taille de son entrée (N). Notre projet se rapprochant du problème du voyageur de commerce, nous pouvons utiliser sa complexité de temps :

\begin{equation}

On peut remarquer que plus les entrées (n) augmentent, plus la division 1/2 est négligeable. C’est pourquoi nous pouvons la synthétiser de cette façon :

\begin{equation}
O(n!)
\end{equation}

Enfin on peut démontrer la classe de complexité de notre problème. Pour commencer nous devons démontrer que notre problème est dans NP. Pour cela, nous allons donc utiliser le problème de décision précédemment établi afin de justifier que celui-ci est facilement vérifiable :

***Existe-t-il un cycle hamiltonien dans G de longueur inférieure à k ?***

Imaginons donc un plan avec des villes et des trajets. On remarque que de ce problème ressort un cycle hamiltonien où les villes sont les différents points par lesquels devront passer les camions. Ainsi, si nous avons en données les différentes distances (ou temps de trajets) entre nos différentes villes nous pouvons facilement vérifier que le cycle hamiltonien de notre plan possède une longueur ou non inférieure à une distance (ou temps de trajets) k.

À la suite de cette démonstration, nous pouvons démontrer que notre problème est NP-Complet ou non. En effet, pour cela nous devons tout d’abord remarquer qu’un problème est NP-Complet « si il est au moins aussi difficile qu’un autre problème NP-Complet ». Sachant que le problème du voyageur de commerce est NP-Complet, on peut donc effectuer une réduction polynomiale afin de rapprocher notre problème de celui du TSP.

En effet, comme dit dans le premier paragraphe de notre partie Complexité, on peut remarquer que le problème du voyageur de commerce pose la même problématique et thématique que celui de notre projet. C’est ainsi que nous pouvons supposer que notre problème est au moins aussi difficile que celui du TSP. Notre problème du projet Data est donc NP-Complet. C’est alors que si on suppose qu’il existe un parcours des camions. On doit donc prouver qu’il existe un parcours du voyageur de commerce. Si on suppose qu’il existe un parcours du voyageur de commerce. On doit donc prouver qu’il existe un parcours des camions.

Version Cycle Hamiltonien :
Nous pouvons également justifier la catégorie NP-Complet de notre problème à l’aide du problème du cycle Hamiltonien. En effet, le cycle hamiltonien est NP-Complet. Or, si nous effectuons une réduction polynomiale de notre projet au niveau d’un cycle hamiltonien, nous pouvons remarquer que les villes de notre plan sont les sommets et que notre camion doit donc passer une et une seule fois par chaque ville (sommet) et revenir à son point de départ, tel un cycle. Nous pouvons donc supposer que notre problème est au moins aussi difficile que celui du cycle hamiltonien. Notre problème du projet Data est donc NP-Complet.

### &emsp; **2.2. Présensation du choix et description de l'algorithme** <a name="2.2"></a>

#### &emsp; &emsp; **2.2.1 Choix d'une solution** <a name="2.2.1"></a>

Étant donné que le problème du voyageur de commerce est un problème largement étudié du fait de sa complexité et du nombre important de problèmes dérivés [[3]](https://interstices.info/le-probleme-du-voyageur-de-commerce/), il existe de nombreux algorithmes heuristiques et métaheuristiques afin de le résoudre.    
Notre problème étant très similaire, nous pouvons nous baser sur l’un de ces algorithmes. Chacun d’eux a ses avantages et ses inconvénients. La méthode retenue ne sera pas la même dans la mesure où l’on privilégie le temps de calcul, la qualité de la solution, ou encore le choix entre plusieurs solutions.

On peut commencer par citer la famille des algorithmes gloutons. Ces algorithmes font le choix de l’optimum local à chaque étape. Voici deux algorithmes de ce type :
*   L’algorithme de l’insertion est un algorithme plutôt simple. Son principe est que le parcours est construit pas à pas en y insérant de nouvelles villes. À un instant donné de l’algorithme, un certain cycle de villes a été construit. L’étape suivante consiste à insérer une ville supplémentaire dans le cycle de manière optimale, c’est-à-dire qu’elle augmente au minimum la longueur totale du cycle. À l’étape initiale de l’algorithme, le parcours de voyageur est composé de deux villes, la ville de départ et celle qui en est la plus proche. L’algorithme se termine lorsque toutes les villes à visiter ont été insérées.
*   L’algorithme du plus proche voisin et également un algorithme simple pouvant être considéré comme glouton. Son principe est de se déplacer de ville en ville en appliquant la règle du plus proche voisin, c’est-à-dire en sélectionnant la prochaine ville tel que le poids entre la ville courante et la prochaine ville soit minimal, et ce jusqu’à avoir visité toutes les villes. Il faut enfin revenir à la première ville choisie pour obtenir un cycle.
 
Il existe beaucoup d’autres algorithmes, comme l’algorithme de descente locale, qui combine un algorithme glouton pour le chemin initial avec un opérateur de voisinage pour générer plusieurs nouvelles solutions voisines. 

Ce **principe de combiner plusieurs algorithmes est intéressant**, il permet d'obtenir plusieurs solutions initiales valides mais pas forcément optimisées, on utilise ensuite un autre algorithme métaheuristique permettant d'améliorer ces solutions pour obtenir un résultat final performant.
Il est notamment utilisé dans la référence [[4]](https://docplayer.fr/2653298-Resolution-du-probleme-du-voyageur-de-commerce-metaheuristique.html) page 13, en effet, afin de résoudre un problème avec 250 villes, ils combinent un algorithme des fourmis, un algorithme génétique, ainsi qu'un algorithme du recuit simulé.


Nous porterons donc notre choix sur la **mise en place d'une combinaison d'algorithmes** pour résoudre notre problème, et plus spécifiquement d'une combinaison de **2 algorithmes**. Nous débuterons d’abord par une phase de développement des premiers algorithmes, sans prendre en compte les différentes contraintes.

Le premier algorithme sera un algorithme glouton, permettant de trouver un trajet passant par toutes les villes et assurant donc les livraisons. Il nous permettra de récupérer une solution valide, mais pas optimale. Pour cet algorithme,**nous avons choisis l'algorithme du plus proche voisin.**

Puis, nous utiliserons un algorithme métaheuristique, qui prendra en entrée la solution précédemment trouvée, et qui essayera de l'améliorer. **Pour celui-ci, nous avons choisi l'algorithme métaheuristique de la recherche tabou.**

Par la suite, nous ajouterons la prise en compte du trafic sur les routes pour les deux algorithmes, il sera représenté par une augmentation de la pondération d’une arête en fonction du nombre de véhicules sur la route. Par exemple, si une route est très embouteillée, on pourra ajouter 2h de trajet pour cette route, elle ne sera donc peut-être plus la route la plus optimale.\
 Nous ajouterons ensuite les contraintes supplémentaires, comme le respect d’une fenêtre de temps de livraison pour un objet.

#### &emsp; &emsp; **2.2.2 Fonctionnement et paramètres** <a name="2.2.2"></a>

**Paramètres :**\
Lors de la réalisation de notre algorithme, nous avons eu besoin d'utiliser un certain nombre de paramètres. Ces variables ont une importance dans le résultat de notre algorithme et nous devons donc les définir.
- Les villes à visiter :\
Ces paramètres sont représentés par une matrice. Celle-ci permet de connaître le temps séparant chaque duo de villes. Elle est générée avec des temps entre chaque ville qui sont aléatoires, avec pour valeur maximale 200 minutes. 
- Le nombre d'itérations:\
Cette valeur correspond au nombre de répétitions de notre recherche taboue. Ainsi nous allons définir un nombre d'itérations maximum. 
- La taille de la liste tabou:\
Cette valeur correspond à la mémoire de la liste tabou et le cerveau principal de cet algorithme d'où la thèse d'Amélie Lambert « Les méthodes de résolution approchées pour la programmation en nombres entiers » [[5]](http://cedric.cnam.fr/~lamberta/ECE/OutLog/OL_Heur_C4_2016-2017.pdf)
En effet si cette taille est courte, les interdictions sont mineures et la recherche suit bien plus les optimums locaux rencontrés. L’algorithme va parcourir des distances assez faibles dans l’espace de recherche et risque donc de tomber dans une boucle, c’est-à-dire retomber sans cesses dans le même optimum local.
Si la taille est assez longue, les interdictions sont plus nombreuses et la recherche risque de manquer de nombreux optimums locaux sur le chemin. L’algorithme va parcourir de plus grandes distances dans l’espace de recherche et permet de réduire le risque de boucle. Ainsi il y a plus d’améliorations de la valeur de la fonction objective.



**Fonctionnement :**

L'idée de la recherche tabou consiste, à partir d'une position donnée, à en explorer le voisinage et à choisir la position dans ce voisinage qui minimise la fonction objectif.
Dans notre cas, une position donnée est un chemin, le voisinage est représenté par l'ensemble des chemins pour lesquels nous avons échangé une paire de sommet, et enfin, la fonction objectif est représenté par le temps de trajet pour parcourir toutes les villes, qui doit être le plus court possible.

Voici un exemple pour comprendre comment le voisinage est calculé.

Pour le chemin **0,1,2,3,0**, les voisins sont :
    
    0,2,1,3,0    
    0,3,2,1,0
    0,1,3,2,0

Voici le principe du fonctionnement de notre algorithme :
1. On part d'un élément de notre ensemble de recherche, qui est donc notre chemin initial lors de la première itération, et l'ancien meilleur voisin lors des itérations suivantes. Ce chemin est déclaré élément courant.
2. On génère le voisinage de l'élément courant et on choisit le meilleur d'entre eux comme nouvel élément courant, parmi ceux absents de la liste tabou. Dans notre cas, le meilleur élément est le chemin le plus court en temps. On ajoute ensuite cet élément à la liste tabou.
3. On boucle ensuite jusqu'à condition de sortie, donc lorsque le nombre d'itération maximal est atteint, et on renvoie la meilleure solution trouvé.

Lors de ce fonctionnement, il faut aussi ajouté le fait qu'on supprime la plus ancienne solution trouvée lorsque la taille de la liste tabou à atteint sa taille maximale

#### &emsp; &emsp; **2.2.4 Modélisation du problème selon le formalisme de l'algorithme** <a name="2.2.4"></a>

Dans cette section vous sera présenté pour commencer la formulation du problème VRP classique sans contraintes puis nous enchainerons avec l'ajout de la contrainte temporelle avec le problème VRPTW (Vehicule Routing Problem with Time Windows) pour finir avec la contrainte de capacité pour la CVRP (Capacitated Vehicule Routing Problem).

##### &emsp; &emsp; &emsp; **2.2.4.1 Modélisation du VRP**

Un graphe $G = (N, A)$ représente notre problème où :


*   $N$ représente les positions des clients et du dépôt,

*    $A$ représente les arcs entre deux clients $i, j ∈ N$.




Pour trouver l’ordre de visite des clients, nous définissons les variables de décisions
comme suit :   

$$
X_{ij} = \left\{
    \begin{array}{ll}
        1 & \mbox{si (i,j) est parcouru par un véhicule,} \\
        0 & sinon,
    \end{array}
\right.
$$   

Il nous est donc possible d’écrire formellement le modèle de VRP


 $Minimiser \sum_{i=1}^n \sum_{j=1}^n c_{ij} x_{ij}$ (1)

Avec les contraintes suiviantes : 

$n$ nombre de clients ( ou sommets)

$c_ij$ le coût de l'arête entre les sommets i et j ( temps de parcours)

$\sum_{i=1}^n X_{ij} = 1 \forall 1 \leq j \leq n$ (2)

$\sum_{j=1}^n X_{ij} = 1  \forall  1 \leq i \leq n$ (3)

$\sum_{i=1}^n \sum_{l=1}^n X_{il} = \sum_{l=1}^n \sum_{j=1}^n$ (4)

$\sum_{j=1}^n X_{0j} = 1$ (5)

$\sum_{j=1}^n X_{i0} = 1$ (6)

###### &emsp; &emsp; &emsp; &emsp; **2.2.4.1.1 Détails des équations**

La fonction objectif (1) signifie que l'objectif de notre problème d'optimisation est de minimiser la somme des coûts de toutes les tournées, et donc par conséquent limiter la durée de toutes les tournées.
   
La formulation du problème nécessite de satisfaire certaines contraintes :   
 -  Les contraintes (2) et (3) imposent que chaque client soit desservi une et une seule fois, et donc que chaque sommet ne soit visité qu'une seule fois.

 - La contrainte (4) indique que notre graphe est complet, et donc que chaque sommet est relié aux autres sommets.

 - Les contraintes (5) et (6)imposent que le point de départ est aussi le point d'arrivé.


##### &emsp; &emsp; &emsp; **2.2.4.2 Modélisation du VRPTW**

Le problème du VRPTW est une variante du problème VRP, il s'agit donc du problème VRP auquel on ajoute une contrainte supplémentaires : chaque client doit fournir une limite de temps  des fenêtre de temps supérieure et inférieure entre lesquelles il désire être
servi.

Il a été montré que le problème VRP classique était NP-difficile, et que ce résultat pouvait être
étendu au VRPTW. Nous formulons le problème VRPTW en utilisant la formulation de VRP classique en y ajoutant les variables, les constantes ainsi que les équations suivantes.

Avec les variables suiviantes à définir :





*   $a_i$ = instant d’arrivée chez le client $i ∈ N$
*   $b_i$ = instant de début de service chez le client $i ∈ N$
* $b^v_0$ = instant auquel le véhicule $v$ quitte le dépôt
* $b^v_{n+1}$ = instant auquel le véhicule $v$ retourne au dépôt
* $w_i$ = temps d’attente chez le client $i ∈ N$



Avec les constantes connues :





*   $e_i$ = borne inférieure de la fenêtre de temps du client $i ∈ N$
*   $l_i$ =  borne supérieure de la fenêtre de temps du client $i ∈ N$
* $c_{ij}$ = coût du déplacement de $i$ à $j$ : $i, j ∈ N$
* $t_{ij}$ = le temps de parcours entre les deux clients $i$ et $j$, $i, j ∈ N$
* $s_i$ = temps de service chez le client $i ∈ N$



Il nous est possible d’écrire les contraintes supplémentaires
de la formulation de VRP pour formuler le problème VRPTW comme suivantes :

$x^v_{ij} = 1 \Rightarrow b_i + s_i + t_{ij} \leq b_j, \forall i, j ∈ N, v ∈ V$ (7)

$x^v_{Oj} = 1 \Rightarrow b^v_0 + t_{0j} \leq b_j, \forall j ∈ N, v ∈ V$ (8)

$x^v_{i,n+1} = 1 \Rightarrow b_i + s_i + t_{i,n+1} \leq b^v_{n+1}, \forall i ∈ N, v ∈ V$ (9)

$e_i \leq b_i \leq l_n, \forall i ∈ N$ (10)

$e_0 \leq b^v_0 \leq l_0, \forall v ∈ V$ (11)

$e_{n+1} \leq b^v_{n+1} \leq l_{n+1}, \forall v ∈ V$ (12)

Le temps d’utilisation réel des véhicules dans la solution (X) est :


$temps horaire(X) = \sum_{v∈V} (b^v_{n+1} - b^v_0)$ (13)

###### &emsp; &emsp; &emsp; &emsp; **2.2.4.2.1 Détails des équations**

Les contraintes (7-12) définissent les contraintes temporelles. La mesure (13), quant à elle, permet d’identifier le temps total d'utilisation des véhicules.

##### &emsp; &emsp; &emsp; **2.2.4.3 Modélisation du CVRP**

Le problème du CVRP (Capacitated Vehicule Routing Problem) est une variante du problème VRP, il s'agit donc du problème VRP auquel on ajoute une contrainte supplémentaire : le nombre de véhicules et la capacité des véhicules.

Soit :



$x_{ijk}$ égale à 1 si le véhicule k parcourt l'arc ($v_i$, $v_j$), noté plus simplement ($i$,$j$)

Nous supposons que le graphe $G = (N,A)$ est complet, et donc que tous les sommets sont reliés entre eux. Où $N={v_1,...,v_n}$ est un ensemble de sommets modélisant les adresses des clients ainsi que le dépôt et $A={(v_i,v_j) : i \leq j; v_i,v_j \in V }$ est un ensemble d'arêtes ou d'arcs qui relient les clients et le dépôt.

Cela signifie qu'une adresse est accessible à partir de n'importe quelle autre adresse.

Les autres constantes du problème sont les suivantes : 

$n$ : Nombre de clients (correspond aux sommets)

$m$ : Nombre de véhicules

$Q$ : Capacité des véhicules

$c_{ij}$ : Coût de l'arête entre les sommets $i$ et $j$ (distance ou temps de parcours)

Les variables de décision du problème sont les $x_{i,j,k}$  évoquées plus haut :

$$x_{ijk} = \left\{
  \begin{array}{rcr}
    1 & si&(i,j)&est&parcouru&par&le&véhicule&k.\\
    0 & sinon\\
  \end{array}
\right.$$


Notre problème s'écrit donc de la façon suivante :

$Minimiser $$\sum_{i=1}^n$$ $$\sum_{j=1}^n c_{ij}$$ $$\sum_{k=1}^m x_{ijk}$$ $ (1)

Avec les contraintes suivantes : 

$ $$\sum_{i=1}^n$$ $$\sum_{j=1}^m x_{ijk}$$ = 1 $    $ \forall 1 \leq j \leq n $ (2)

$ $$\sum_{j=1}^n$$ $$\sum_{k=1}^m x_{ijk}$$ = 1 $    $ \forall 1 \leq i \leq n $ (3)

$ $$\sum_{i=1}^n$$ $$\sum_{l=1}^m x_{ilk}$$ = $$\sum_{l=1}^n$$ $$\sum_{j=1}^m x_{ljk}$$ $ (4)

$ $$\sum_{j=1}^n x_{0jk}$$ = 1 $    $ \forall 1 \leq k \leq m $ (5)

$ $$\sum_{i=1}^n x_{i0k}$$ = 1 $    $ \forall 1 \leq k \leq m $ (6)

$ $$\sum_{i=1}^n$$ $$\sum_{j=1}^n x_{ijk}$$  \leq Q$    $ \forall 1 \leq k \leq m $ (7)

$x_{i,j,k} \in 0,1$  $\forall 0 \leq i,j \leq n;$  $1\leq k \leq m $ (8)



###### &emsp; &emsp; &emsp; &emsp; **2.2.4.3.1 Détails des équations**

Sous cette formulation, (1) signifie que l'objectif du problème d'optimisation est de minimiser la somme des coûts de toutes les tournées. 

La formulation du problème nécessite de satisfaire certaines contraintes :   
 - Les contraintes (2) et (3) imposent que chaque client soit desservi une et une seule fois. 
 - La contrainte (4), quant à elle, assure la conservation de flot. 
 - La contrainte (5) assure que chaque tournée commence et se termine au dépôt. 
 - Les contraintes (6) et (8) sont les contraintes de capacité et d'intégrité.
 - La contrainte (7) est la contrainte de binarité sur les variables de décision $x_{i,j,k}$.



##### &emsp; &emsp; &emsp; **2.2.4.4. Choix de la modélisation**

On peut en déduire qu'utiliser une liste tabou trop grande n'est pas forcément bénéfique. Une grande liste tabou permet en effet de pouvoir sortir des optimals locaux et ainsi d'éviter les cycles, mais elle peut aussi rater plusieurs optimals locaux du à son nombre important d'interdictions. A contrario, si la taille de la liste tabou est plus petite, la recherche épouse mieux les optimals locaux rencontrés, cependant, l’algorithme tend à parcourir de moins grandes distances dans l’espace de recherche, et le risque de cycles est plus grand. Comme vu dans la référence [[9]](https://leeds-faculty.colorado.edu/glover/227%20-%20A%20Users%20Guide%20to%20Tabu%20Search%20-%20conversion%203%2026%2014.pdf), partie 3.2.3, les tailles de listes tabou qui fournissent de bons résultats augmentent souvent avec la taille du problème, c'est pourquoi une liste tabou faisant la taille du nombre de villes est intéressante dans notre cas.
